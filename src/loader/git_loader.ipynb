在这个demo中,我们使用GitLoader从本地Git仓库中加载了RAG-Client这个项目的所有代码

```python
import os
import sys

from pathlib import Path

project_root = Path(os.getcwd()).parent.parent.resolve()

sys.path.append(str(project_root))
```

```python
from git import Repo

repo = Repo.clone_from(
    "http://192.168.99.60:4000/BizEditor/RAG-Client.git", to_path="./example_data/test_repo1"
)
branch = repo.head.reference
repo
```

```python
from langchain_community.document_loaders import GitLoader
loader = GitLoader(repo_path="./example_data/test_repo1/", branch=branch)
data = loader.load()

print(len(data))
data
```

引入SemanticChunker语义分词器

```python
from langchain_experimental.text_splitter import SemanticChunker
from common import get_embedding

ollama_embeddings = get_embedding()

text_splitter = SemanticChunker(ollama_embeddings)
```

```python
texts = text_splitter.split_documents(data)
print(len(texts))
texts
```
